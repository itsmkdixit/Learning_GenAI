##### Stemming
- Process of reducing a word to its word stem that affixes to duffixes and prefixes or to the roots of words known as a lemma.
- It is important in natural langunge understanding (NLU) and processing (NLP)

_Let's see an example_ :- 
- Imagine you are solving a classification problem where you are doing a sentiment analysis of a product, essentially to figure out the postive and negative reviews.
- Now the reveiws have words like work, working, worked or fail, fails, falied, failing.
- Work and Fail are the stem words. 
- So instead of increasing the number of input features, we can use these stems words that does not take away the essense of the problem I am trying to solve.

```python
#Let's see an example - PorterStemmer

words = ["worked", "works", "working" , "failed" , "failing", "fails", "programming", "programmed", "programs"]
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words:
    print(word+"--->"+stemming.stem(word))
```

```python
#Stemming disadvantages - changes the meaning

print(stemming.stem('history'))
print(stemming.stem('congratulations'))
print(stemming.stem('habitual'))
```

```python
#Regex Stemmer

from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|n$|able$|ly$', min = 4)

print(reg_stemmer.stem('eating'))
print(reg_stemmer.stem('heavenly'))
print(reg_stemmer.stem('understandable'))
print(reg_stemmer.stem('shaken'))

```

```python
#Snowball Stemmer - better accuracy

from nltk.stem import SnowballStemmer
sstemmer = SnowballStemmer('english')
print(words)
```

```python
for word in words:
    print(word +"-->"+sstemmer.stem(word))
```

```python
print(sstemmer.stem('history'))
print(sstemmer.stem('congratulations'))
print(sstemmer.stem('habitual'))
```

```python
#But let's see where it performs better than Porterstemmer

stemming.stem('fairly'), stemming.stem('manly')
```

```python
sstemmer.stem('fairly'), sstemmer.stem('manly')
```
