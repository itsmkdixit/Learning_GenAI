##### Lemmatization
- We will get the root word instead of the stem word in Stemming. The root word is also called 'lemma'.
- WordNetLemmatizer class in NLTK is a thin wrapper around the wordnet corpus. This class used the morphy() function to the WordNet CorpusReader to find a lemma

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
#using Postags(pos) [Noun - 'n' (default), verb - 'v', adjective - 'a' , adverb - 'r']
lemmatizer.lemmatize("going"), lemmatizer.lemmatize("going", pos = 'v'), lemmatizer.lemmatize("irritating", pos = 'v')
```
