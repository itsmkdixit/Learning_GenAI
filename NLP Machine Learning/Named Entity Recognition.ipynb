Named Entity Recognition(NER) is identifying and classifying real-world entities in text. These usually fall into the following category:
- PERSON - people
- ORG - comapnies, institutions
- GPE - countries, cities, locations
- DATE/TIME
- MONEY
- PRODUCT
- EVENT
- etc.

We use named entity chunker to chunk the given list of tagged tokens _(nltk.ne_chunk)_

```python
# Let's see an example

text = 'My name is Mayank and I live in Bangalore since 2016, which is one of the most beuatiful city in the Southern part of India.'

import nltk

nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("maxent_ne_chunker")
nltk.download("words")

tokens = nltk.word_tokenize(text) #sentence to word
postag = nltk.pos_tag(tokens) #POS tagging

entities = nltk.ne_chunk(postag)
print(entities)
```
