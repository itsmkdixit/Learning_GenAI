Bag of Words represents text by word counts, ignoring grammar and word order.

Example:
- S1: "fraud detected in transaction"
- S2: "transaction fraud fraud detected"

BoW treats both as almost the same, because:
- same words
- similar counts
- order doesn’t matter

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "fraud detected in transaction",
    "transaction fraud fraud detected"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
#Checking on an actual spam file

import pandas as pd

messages = pd.read_csv('/Users/mayankdixit/Documents/AI_Learning_Conda/NLP Machine Learning/spam.csv',encoding = 'latin-1')
```

```python
messages = messages[['v1', 'v2']]
messages.columns = ['label', 'message']
```

```python
messages
```

```python
#Data Preprocessing

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
corpus = []

for message in messages['message']:

    # 1. Remove non-alphabetic characters
    text = re.sub('[^a-zA-Z]', ' ', message)

    # 2. Convert to lowercase
    text = text.lower()

    # 3. Split into words
    words = text.split()

    # 4. Remove stopwords
    filtered_words = []
    for word in words:
        if word not in stop_words:
            filtered_words.append(word)

    # 5. Apply stemming
    stemmed_words = []
    for word in filtered_words:
        stemmed_words.append(ps.stem(word))

    # 6. Rebuild sentence
    clean_text = ' '.join(stemmed_words)

    # 7. Add to corpus
    corpus.append(clean_text)
```

```python
corpus
```

```python
# Creating BoW Model

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500) # only top 2500 features will be taken. binary = True for binary BoW
X = cv.fit_transform(corpus).toarray()
y = messages['label']

print("BoW shape:", X.shape)
print("Example features:", cv.get_feature_names_out()[:20])
```

##### N Gram -  n-gram is a sequence of n consecutive tokens.
    - Unigram (1-gram) → single words
    - Bigram (2-gram) → two consecutive words
    - Trigram (3-gram) → three consecutive words

So why does N- gram exists?

The Bag of Words ignore the word order which creates dangerous cases like:
- fraudulent
- not fraudulent

Unigram treat these as almost identical, but Bigram fix this by capturing the local context
- "not fraudulent" becomes a disctinct feature
- "fraudulent" alone no longer dominates

So the purpose of "n-grams" is to capture the meaning theat comes from word combinations.


```python
#Let's see an example 

#Unigrams - these are deafult params inside CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(ngram_range=(1, 1))
X = cv.fit_transform(corpus)
```

```python
cv.vocabulary_
```

```python
# Unigrams + Bigrams

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(ngram_range=(1, 2))
X = cv.fit_transform(corpus)
```
