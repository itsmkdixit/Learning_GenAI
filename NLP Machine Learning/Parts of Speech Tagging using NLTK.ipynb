##### POS Tagging = Assigning each word a grammatical role (noun, verb, adjective, etc.).
- WordNet Lemmatizer needs POS to work properly.

- "better" →

  - adjective → good

  - noun → better

- "running" →

  - verb → run

  - noun → running

```python
#Let's see how this improves lemmatization

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# ----------------------------------------------------
# DOWNLOAD REQUIRED NLTK RESOURCES (only run once)
# punkt           → tokenizers
# averaged_perceptron_tagger → POS tagger
# wordnet         → lexical database used for lemmatization
# ----------------------------------------------------
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")

# ----------------------------------------------------
# SAMPLE PARAGRAPH (MULTI-SENTENCE TEXT)
# ----------------------------------------------------
paragraph = """
Natural Language Processing is transforming how machines understand human communication.
People are creating powerful applications using text data.
This progress will accelerate in the coming years.
"""

# ----------------------------------------------------
# STEP 1: SPLIT PARAGRAPH INTO SENTENCES
# ----------------------------------------------------
sentences = nltk.sent_tokenize(paragraph)

# Initialize the WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# ----------------------------------------------------
# FUNCTION TO MAP NLTK POS TAGS → WORDNET POS TAGS
# NLTK uses "NN", "VBZ", etc.
# WordNet uses NOUN, VERB, ADJECTIVE, ADVERB constants.
# This function converts between the two.
# ----------------------------------------------------
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ      # adjective
    elif treebank_tag.startswith('V'):
        return wordnet.VERB     # verb
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN     # noun
    elif treebank_tag.startswith('R'):
        return wordnet.ADV      # adverb
    else:
        return wordnet.NOUN     # default (common fallback)

# ----------------------------------------------------
# STEP 2: PROCESS EACH SENTENCE
# ----------------------------------------------------
processed_sentences = []

for sent in sentences:

    # Tokenize the sentence into words
    tokens = nltk.word_tokenize(sent)

    # Apply POS tagging (assign grammatical roles)
    pos_tags = nltk.pos_tag(tokens)

    # ----------------------------------------------------
    # STEP 3: LEMMATIZE EACH WORD BASED ON ITS POS TAG
    # Logic:
    #   - extract (word, pos_tag)
    #   - convert NLTK POS tag → WordNet POS tag
    #   - lemmatize using correct POS
    #   - remove punctuation by checking isalpha()
    # ----------------------------------------------------
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(pos))
        for word, pos in pos_tags
        if word.isalpha()   # skip punctuation
    ]

    # Rebuild cleaned sentence
    processed_sentences.append(" ".join(lemmatized_words))

# Final processed output
processed_sentences
```

```python
# Practicing more assignments - Extract only nouns from the paragraph

import nltk
from nltk import pos_tag, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Downloads
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")

lemmatizer = WordNetLemmatizer()

def extract_nouns(text):
    # Tokenize text
    tokens = word_tokenize(text)
    
    # POS tagging
    pos_tags = pos_tag(tokens)
    
    nouns = []
    
    for word, tag in pos_tags:
        if tag in ["NN", "NNS", "NNP", "NNPS"]:
            # Lemmatize nouns specifically as NOUN
            lemma = lemmatizer.lemmatize(word, pos=wordnet.NOUN)
            nouns.append(lemma)
    
    return nouns


# Example
text = """
Natural Language Processing enables machines to understand human language.
Developers use models, datasets, and algorithms to analyze text.
"""

print(extract_nouns(text))
```
